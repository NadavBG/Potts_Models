```python
%load_ext autoreload
%autoreload 2

import numpy as np
from pathlib import Path
import SBM
import SBM.utils.utils as ut
import SBM.utils.utils_plot as up

ROOT = Path(SBM.__file__).resolve().parents[2]
data_dir = ROOT / "data"
results_dir = ROOT / "results"
```

```python
# Example without test set
file = (
    results_dir
    / "CM/CM_ModelSBM_N_chains70_N_iter400_Param_initzero_k_MCMC100000_lambda_J0.0_lambda_h0_m1_theta0.3_N_Av1_R0.npy"
)
output = np.load(file, allow_pickle=True)[()]
align_mod = ut.Create_modAlign(
    output,
    output["align"].shape[0],
    delta_t=output["options0"]["k_MCMC"],
    temperature=1,
)
output["align_mod"] = align_mod
```
